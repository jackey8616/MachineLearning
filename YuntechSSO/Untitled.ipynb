```python
from pytesseract import image_to_string
from PIL import Image
from IPython.core.display import Image as Img, display

import re
from os import listdir
from os.path import isfile, join

def imageToString(img):
    string = image_to_string(img)
    #string = re.sub('\W', '', string)
    return string

def imageToStringLang(img):
    string = image_to_string(img, lang="fontsso")
    #string = re.sub('\W', '', string)
    return string
    

def defaultLang():
    for each in listdir('./TIFF_Inv'):
        if(each.endswith('.tif')):
            image = Image.open('./TIFF_Inv/' + each)
            before = imageToString(image)
            after = imageToStringLang(image)
            if before != after:
                display(image)
                print('Before: %s, After: %s' % (before, after))
```

```python
defaultLang()
```
